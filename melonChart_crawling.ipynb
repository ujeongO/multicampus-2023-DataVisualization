```python
# 순위 변동 표시 함수 정의
def view_rank_change(rank_change):
    condition = rank_change[:5]

    if condition == "순위 동일":
        rank_change = '-'
    elif condition == "단계 상승":
        rank_change = '⬆️'+rank_change[-1]
    elif condition == "단계 하락":
        rank_change = '⬇️'+rank_change[-1]
    else:
        rank_change

    return rank_change
```

```python
# 차트 전체 정보 가져오기
song_data = []
rank = 1  # 초기화

for i in range(100):
    # 순위 변동
    rank_change = soup.select("tbody>tr>td>div.wrap>span.rank_wrap")[i].text.replace('\n', '')
    
#     순위변동 함수 왜 안 먹는겨?.?
#     view_rank_change(rank_change)

    condition = rank_change[:5]

    if condition == "순위 동일":
        rank_change = '-'
    elif condition == "단계 상승":
        rank_change = '⬆️'+rank_change[-1]
    elif condition == "단계 하락":
        rank_change = '⬇️'+rank_change[-1]
    else:
        rank_change
    
        
    # 곡명
    songs = soup.select("tbody>tr>td>div .wrap_song_info>div.ellipsis.rank01")[i].text.replace('\n', '')
    # 가수명
    singers = soup.select("tbody>tr>td>div .wrap_song_info>div.ellipsis.rank02>span")[i].text
    # 앨범명
    albums = soup.select("tbody>tr>td>div .wrap_song_info>div.ellipsis.rank03>a")[i].text
    # 좋아요 수
    likes = soup.select("tbody>tr>td>div>button>span.cnt")[i].text[5:]
    
    # mylist로 묶기
    mylist = ["Melon", rank, rank_change, songs, singers, albums, likes]
    song_data.append(mylist)
    # 순위
    rank += 1
    
```

```python
# data frame 생성

pd.DataFrame(song_data, columns=["서비스업체", "순위", "순위 변동", "곡명", "가수", "앨범명", "좋아요 수"])
# pd.DataFrame(song_data, columns=["서비스업체", "순위", "곡명", "앨범명", "좋아요 수"])
```

```python
# csv 파일로 저장
df = pd.DataFrame(song_data, columns=["서비스업체", "순위", "순위 변동", "곡명", "가수", "앨범명", "좋아요 수"])
# path = /Users/oyujeong/Desktop/mulcam_ujeong/data'
# file name = melon_ranking_20231019.csv
df.to_csv('/Users/oyujeong/Desktop/mulcam_ujeong/data/melon_ranking_crawling.csv', index = False)

df
```
